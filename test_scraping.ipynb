```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time
```

```python
driver = webdriver.Chrome("/Users/abhishekpeiris/Downloads/chromedriver")
driver.get("https://ikman.lk/en/ads/sri-lanka/electronics")
time.sleep(4)
```

```python
items_list = []

for x in range(10):
    print(x)
    time.sleep(5)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")

    page_source = driver.page_source
    soup = BeautifulSoup(page_source, features="html.parser")


    items = soup.findAll('li', attrs = {'class':'normal--2QYVk gtm-normal-ad'})

    for item in items:
        description = item.find('div', attrs = {'class':'description--2-ez3'}).text
        price = item.find('div', attrs = {'class':'price--3SnqI'}).span.text
        name = item.h2.text
        # print(name, " | " , description, " | ", price)
        items_list.append({"name": name, "description" : description, "price" : price})

    driver.find_element(By.CLASS_NAME, "gtm-next-page").click()
    time.sleep(2)
```
